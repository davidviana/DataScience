#Lendo DataFrame

```python
df_banco = pd.read_csv('Banco.csv')
df_banco.head()
```

DataSet

```python
display(df_banco.describe())
```

Tipo de Dados

```python
df_banco.dtypes
```

Dados Nulos

```python
display(df_banco.info())
```

Colunas (Series)

```python
series = df_banco['balance'].min()
print(series)
```

Dropando Series

```python
df_drop = df_banco.drop(axis=1, labels=['job', 'education'])
display(df_drop)
```

## Linha

```python
display(df_banco.tail(2))
```

Iterando sobre as ocorrencias (linhas)

```python
for a,b in df_banco.iterrows():
  print(a)
  print(b)
```

# Subsets

Selecionando linhas - iloc

```python
df_banco.iloc[0:10,[1,2]] #primeiro param é linha, segundo param column
```

Selecionando linhas - loc

```python
display(df_banco.loc[(df_banco['marital'] == 'single') & (df_banco['job'] == 'unemployed')])
```

Queries

```python
display(df_banco.query('balance > 4000'))
```
